```python
# Create new columns - mins per match
epl['MinsPerMatch'] = (epl['Mins']/epl['Matches']).astype(int)
epl['GoalsPerMatch'] = (epl['Goals']/epl['Matches']).astype(float)
epl.head()
```

```python
# Total Goals - for the season
Total_goals = epl['Goals'].sum()
Total_goals
```

```python
# Peanlty Goals
Total_penalty_goals = epl['Penalty_Goals'].sum()
Total_penalty_goals
```

```python
# Penalty Attempts
Total_penalty_attempts = epl['Penalty_Attempted'].sum()
Total_penalty_attempts
```

```python
# Pie charts for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_not_scored = epl['Penalty_Attempted'].sum() - Total_penalty_goals 
data = [pl_not_scored,Total_penalty_goals]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette("Blues")
plt.pie(data, labels = labels, colors = color, autopct='%.0f%%')
```

```python
# Most players from which Country
nationality = epl.groupby('Nationality').size().sort_values(ascending = False)
nationality.head()
nationality.head(10).plot(kind ='bar', figsize=(12,6), color = sns.color_palette("Blues"))
```
